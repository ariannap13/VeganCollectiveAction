```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Collective action data analysis

```python
df = pd.read_csv('./results/collective_action_features_comments.csv')
```

### Define order of narrative types

```python
df = df.sort_values('narrative')
```

### Plot

```python
color_dict = {"group_0": plt.cm.viridis(0.0),
                "group_1": plt.cm.viridis(0.25),
                "group_2": plt.cm.viridis(0.5),
                "group_3": plt.cm.viridis(0.75),
                "self_0": plt.cm.inferno(0.4),
                "self_1": plt.cm.inferno(0.55),
                "self_2": plt.cm.inferno(0.7),
                "self_3": plt.cm.inferno(0.85)}
             

# plot boxplot
plt.figure()
# plot average, use fixed order for the narrative labels, add edges to the bars. Make the bars wider and dots bigger. Order the bars by narrative order
sns.lineplot(x="narrative", y="collective_action rfreq", data=df, marker='o', markersize=12, linestyle='', err_style='bars', hue='narrative', errorbar=('ci', 95), err_kws={'capsize': 4, 'elinewidth': 2, 'capthick': 2}, sort=False, palette=color_dict)
# remove legend 
plt.legend([],[], frameon=False)
# remove x label
plt.xlabel('')
# change y label to what is in parenthesis
feature_name = 'Collective action'
plt.ylabel(feature_name + ' (%)', fontsize=16)

# set ylabel size
plt.yticks(fontsize=14)
plt.xticks(fontsize=12)
# change x ticks labels to narrative interpretation by using the mapping defined above
plt.xticks(np.arange(8), ["Educate\n(inform)", "Educate\n(inspire)", "Fight\n(protest)", "Fight\n(convert)",
              "Good health\n(personal)", "Choose\nc(personal)", "Good health\n(benefits)", "Choose\n(discuss)"])

# remove all spines except the left one
sns.despine(left=False, bottom=False, right=True)

# rotate x ticks labels
plt.xticks(rotation=47)

# save as svg
plt.savefig('./pics/collective_action.svg', bbox_inches='tight')
```
